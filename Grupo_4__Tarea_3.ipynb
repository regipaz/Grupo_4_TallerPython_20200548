```python
## pregunta 5
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
#   Parámetros de la distribución normal estándar
mu = 0
sigma = 1

#   Generamos una muestra aleatoria de 1000 elementos de una distribución normal estandar
sample = stats.norm.rvs(loc=mu, scale=sigma, size=1000)
```

```python
## pregunta 6
#   Calculamos la media, desviación estándar y mediana
mean = np.mean(sample)
std_dev = np.std(sample)
median = np.median(sample)

#   Mostramos los resultados. Presentamos con 3 decimales
print(f"Media: {mean:.3f}")
print(f"Desviación estándar: {std_dev:.3f}")
print(f"Mediana: {median:.3f}")
```

```python
# Pregunta 7:

#   Graficamos el histograma de la muestra

plt.figure(figsize=(10, 6))
plt.hist(sample, bins=30, density=True, alpha=0.6, color='green', edgecolor='black')

#   Modelamos la curva de la densidad de la distribución normal estándar, en el rango de valores que ocupa la muestra.
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 250)
dens = stats.norm.pdf(x, mu, sigma)

#   Superponemos la curva de la densidad de la distribución normal.
plt.plot(x, dens, 'k', linewidth=2)
plt.title('Histograma de datos generados y curva de distribución normal')
plt.xlabel('Valor')
plt.ylabel('Densidad de Probabilidad')

# Mostramos la gráfica
plt.show()
```
