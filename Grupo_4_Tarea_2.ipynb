```python
import pandas as pd
import matplotlib.pyplot as plt
```

1. Copia el siguiente diccionario en una c√©dula de Colab. A partir de √©l, crea un dataframe llamado ùíèùíáùíê utilizando Pandas.

```python
data_ciudades = {
           'Ciudad': ['Lima', 'Arequipa', 'Trujillo', 'Cusco', 'Chiclayo', 'Piura', 'Iquitos', 'Huancayo', 'Tacna', 'Pucallpa'],
    'Habitantes': [1047996, 100169, 92331, 428450, 305717, 484475, 441649, 385098, 294395, 283734],
    'Area_km2': [2672.28, 1545.77, 1487.7, 116.5, 279.89, 6217.26, 368.9, 109.19, 59.4, 483.44],
    'Altitud_m': [154, 2325, 34, 3399, 29, 29, 106, 3271, 562, 156],
    'Densidad_poblacion_por_km2': [3924, 64.8, 62.1, 3673, 1091.6, 77.9, 1196, 3528.6, 4966, 587.2]
        }

nfo = pd.DataFrame(data_ciudades)
nfo
```

2. . Realiza el mismo procedimiento con el siguiente diccionario. A partir de √©l, crea un dataframe llamado ùíèùíêùíéùíÉùíìùíÜùíî.

```python
data_ciudades_2 = {
          'Ciudad': ['Lima', 'Arequipa', 'Trujillo', 'Cusco', 'Chiclayo', 'Piura', 'Iquitos', 'Huancayo', 'Tacna', 'Pucallpa'],
    'Gentilicio': ['Limense', 'Arequipe√±o', 'Trujillano', 'Cusque√±o', 'Chiclayano', 'Piurano', 'Iquite√±o', 'Huanca√≠no', 'Tacne√±o', 'Pucallpino'],
    'Provincia': ['Lima', 'Arequipa', 'Trujillo', 'Cusco', 'Chiclayo', 'Piura', 'Maynas', 'Huancayo', 'Tacna', 'Coronel Portillo'],
    'Region': ['Lima', 'Arequipa', 'La Libertad', 'Cusco', 'Lambayeque', 'Piura', 'Loreto', 'Jun√≠n', 'Tacna', 'Ucayali']
        }

nombres = pd.DataFrame(data_ciudades_2)
nombres
```

3. Con ambos dataframes, realiza un inner join. Guarda el resultado en un nuevo dataframe
llamado ùíÑùíñùíÇùíÖùíìùíê_1

```python
cuadro_1=pd.merge(nfo, nombres, on='Ciudad', how='inner')
cuadro_1
```

4. Resume las estad√≠sticas descriptivas del dataframe y responde ¬øCu√°l es la m√≠nima densidad poblacional? ¬øA cu√°l ciudad corresponde? ¬øY la m√°xima?



```python
cuadro_1.describe()
```

Por otro lado, la m√°xima densidad poblacional es de 4966 personas por km¬≤, correspondiente a la ciudad de Tacna.

5. Realizar un gr√°fico de barras donde se vea la cantidad de habitantes en cada ciudad



```python
plt.figure(figsize=(10, 6))
plt.bar(cuadro_1['Ciudad'], cuadro_1['Habitantes'], color='green')
plt.xlabel('Ciudad')
plt.ylabel('Habitantes')
plt.title('N√∫mero de Habitantes por Ciudad en Per√∫')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Mostrar el gr√°fico
plt.show()
```

6. Realizar un gr√°fico de dispersi√≥n entre la altura y el n√∫mero de habitantes en las ciudades del dataframe

```python
plt.figure(figsize=(10, 6))
plt.scatter(cuadro_1['Altitud_m'], cuadro_1['Habitantes'], color='red', edgecolor='black')
plt.xlabel('Altitud (m)')
plt.ylabel('N√∫mero de Habitantes')
plt.title('Relaci√≥n entre Altitud y N√∫mero de Habitantes de las Ciudades en Per√∫')

plt.grid(True, linestyle='--', color='grey', alpha=0.7)

plt.tight_layout()
```
